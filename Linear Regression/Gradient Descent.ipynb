```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-4,2,100)
```

```python
def fn(X):
    return 3*X**2 + 4*X -5
Y = fn(X)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
def dy_dx(X): #Gradient
    return 6*X+4
```

```python
alpha = 0.1 #StepSize
Xs = [-3]
Xj = -3
epochs = 20
for i in range(epochs):
    Xj = Xj - alpha*dy_dx(Xj)
    Xs.append(Xj)
```

```python
plt.plot(X,Y)
plt.plot(Xs,fn(np.array(Xs)),c='r')
plt.show()
```
